```
# Read from Tourism Bureau data and parsing the data
# For Indonesian tourist's Halal restaurants
# List them
import requests
from bs4 import BeautifulSoup

url = "https://qutaipei.com/data/TaipeiHRestaurant.html"
get_url = requests.get(url)
get_text = get_url.text
soup = BeautifulSoup(get_text, "html.parser")
list1 = [] 
company = soup.find_all('h4')

for s in company:
  list1.append(s.text) 
  
del list1[-1]
list1
```

```
# list the restaurants' addresses
b= soup.find_all('strong')
list2 = [] 

for c in b:
    list2.append(c.text) 

list2
```

```
# Merge the 2 lists into a table
import pandas as pd
percentile_list = pd.DataFrame(
    {'Restaurant': list1,'Address': list2})

percentile_list
```
